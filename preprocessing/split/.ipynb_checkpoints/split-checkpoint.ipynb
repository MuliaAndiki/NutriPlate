```python
import os
import shutil
import random
from tqdm import tqdm
```

```python

DATA_DIR = "/home/muliaandiki/project/NutriPlate/preprocessing/data/Nutritionverse/train"

IMG_DIR = os.path.join(DATA_DIR, "images")
LBL_DIR = os.path.join(DATA_DIR, "labels")

os.makedirs(IMG_DIR, exist_ok=True)
os.makedirs(LBL_DIR, exist_ok=True)

SPLIT = {
    "train": 0.7,
    "val": 0.2,
    "test": 0.1
}
```

```python
for split in SPLIT.keys():
    os.makedirs(os.path.join(IMG_DIR, split), exist_ok=True)
    os.makedirs(os.path.join(LBL_DIR, split), exist_ok=True)
```

```python
images = [
    f for f in os.listdir(IMG_DIR)
    if f.endswith(".jpg") and os.path.isfile(os.path.join(IMG_DIR, f))
]

random.shuffle(images)

total = len(images)
train_end = int(SPLIT["train"] * total)
val_end = train_end + int(SPLIT["val"] * total)

train_imgs = images[:train_end]
val_imgs = images[train_end:val_end]
test_imgs = images[val_end:]
```

```python
def move_files(file_list, split):
    for img in tqdm(file_list, desc=f"Moving {split}"):
        src_img = os.path.join(IMG_DIR, img)
        src_lbl = os.path.join(LBL_DIR, img.replace(".jpg", ".txt"))

        dst_img = os.path.join(IMG_DIR, split, img)
        dst_lbl = os.path.join(LBL_DIR, split, img.replace(".jpg", ".txt"))

        shutil.move(src_img, dst_img)
        if os.path.exists(src_lbl):
            shutil.move(src_lbl, dst_lbl)

move_files(train_imgs, "train")
move_files(val_imgs, "val")
move_files(test_imgs, "test")
```

```python
print("IMG_DIR exists:", os.path.exists(IMG_DIR))
print("Isi IMG_DIR:")
print(os.listdir(IMG_DIR))
```
